## Tracking a ball through air

```python
import math
from math import radians,sin,cos
from ball_trajectory import Ball_2D
import matplotlib.pyplot as plt
```

```python
#To create a trajectory starting at (0,15) with an initial velocity of 60 meters per second launched at 65 degrees is
traj = Ball_2D(0,15,65)
def test_ball_vacuum(noise):
    x = 0; y = 15
    ball = Ball_2D(x,y,65,60,noise = noise)
    t = 0; dt = 0.25
    while y >= 0:
        x,y = ball.step(dt)
        t += dt
        if y >= 0:
            plt.scatter(x,y)
    plt.axis('equal')
    plt.show()
```

```python
test_ball_vacuum([0,0])
test_ball_vacuum([1,1])
```
